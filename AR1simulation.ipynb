```python
import numpy as np
import matplotlib.pyplot as plt
#np.random.seed(0)
```

```python
T = 500
ϵ = np.random.randn(T,1)

plt.hist(ϵ,100)

vare = np.var(ϵ)
print(vare)
```

```python
x = np.zeros((T,1))

ρ = 0.9 
σ = 0.1

for i in range(T-1):
    x[i+1]=ρ*x[i]+σ*ϵ[i+1]

plt.plot(x)
```

```python
IRF=np.zeros((T,1))
shock = np.zeros((T,1))
shock[10] = 3 # we shock the economy in period 10

for i in range(0,T-1):
    IRF[i+1]=ρ*IRF[i]+σ*shock[i+1]

plt.plot(IRF[:80])
```
